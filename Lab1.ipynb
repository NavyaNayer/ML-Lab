# Importing  Data and Performing Data Visualization Task

Construct various type of plots/charts like histogram, bar chart, pie-chart and scatter plot by importing data from a CSV format file. Further label different axes and data in a plot.

***Import Required Libraries***

The first step is to import the required libraries like pandas and matplotlib. We will be using pandas to read the CSV file and matplotlib to plot different types of charts.

```python
import random
import pandas as pd
import matplotlib.pyplot as plt
```

***Generate new dataset using 'random' package***

```python
data = {
    'age': [random.randint(20, 60) for _ in range(100)],
    'gender': [random.choice(['Male', 'Female']) for _ in range(100)],
    'income': [random.randint(20000, 100000) for _ in range(100)]
}

# Convert data to a pandas dataframe and save to CSV file
df = pd.DataFrame(data)
df.to_csv('data.csv', index=False)
```

**Note:** 'age': [random.randint(20, 60) for _ in range(100)] is a Python list that generates a list of 100 random integer values between 20 and 60 (inclusive) for the 'age' column in our dataset.

***Read CSV File***

The next step is to read the data from the CSV file using pandas. For this example, we will be using a CSV file named "data.csv" created in the previous cell.

```python
data = pd.read_csv('data.csv')
```

# Histogram

A histogram is a chart that shows the distribution of a numerical variable. We can use the **hist()** function of matplotlib to plot a histogram.


```python
plt.hist(data['age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

In this example, we are plotting the age distribution of the dataset. We first call the hist() function and pass the age column of the dataset as the argument. We then label the x and y axes using the xlabel() and ylabel() functions, respectively. Finally, we add a title to the chart using the title() function and display the chart using the show() function.

***Change color of bins***

```python
plt.hist(data['age'],color='green')
# plt.hist([mens_age, female_age], color=['Black', 'Red'], label=['Male', 'Female'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

***Adjust number of bins***

```python
plt.hist(data['age'],color='green',bins = 20)
# plt.hist([mens_age, female_age], color=['Black', 'Red'], label=['Male', 'Female'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

# Bar Chart

A bar chart is a chart that shows the comparison between different categories. We can use the **bar()** function of matplotlib to plot a bar chart.


```python
plt.bar(data['gender'].unique(), data['gender'].value_counts())
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Comparison')
plt.show()
```

In this example, we are plotting the comparison between the genders in the dataset. We first call the bar() function and pass the unique gender categories as the x-axis values and the count of each category as the y-axis values. We then label the x and y axes using the xlabel() and ylabel() functions, respectively. Finally, we add a title to the chart using the title() function and display the chart using the show() function.

***Adjust bar width and figure size***

```python
plt.figure(figsize=(1,2))
# creating Bar chart with a width size of 0.4
plt.bar(data['gender'].unique(), data['gender'].value_counts(),width = 0.4)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Comparison')
plt.show()
```

***Plot horizontal bar chart***

We can use **barh()** function to plot horizontal bars

```python
plt.figure(figsize=(2,1))
# creating Bar chart with a height of 0.4
plt.barh(data['gender'].unique(), data['gender'].value_counts(),height=0.4)
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Comparison')
plt.show()
```

***Add Data Labels***

The **bar_label()** function allows us to provide a number of keyword arguments. This makes plots easy to read and saves our time.

```python
plt.figure(figsize=(2,1))
# creating Bar chart with a height of 0.4
bars = plt.barh(data['gender'].unique(), data['gender'].value_counts(),height=0.4)
plt.bar_label(bars)

plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Comparison')
plt.show()
```

# Pie Chart

A pie chart is a chart that shows the proportion of each category. We can use the **pie()** function of matplotlib to plot a pie chart.


```python
plt.pie(data['gender'].value_counts(), labels=data['gender'].unique())
plt.title('Gender Proportion')
plt.show()
```

In this example, we are plotting the proportion of each gender in the data. We first call the pie() function and pass the count of each gender category as the values and the unique gender categories as the labels. We then add a title to the chart using the title() function and display the chart using the show() function.

***Add labels in piechart***

We can add numeric value in a pie chart by using **autopct** parameter. This parameter allows us to display the percent value using string formatting.

```python
plt.pie(data['gender'].value_counts(), labels=data['gender'].unique(),autopct='%1.1f%%')
plt.title('Gender Proportion')
plt.show()
```

# Scatter Plot

A scatter plot is a chart that shows the relationship between two numerical variables. We can use the **scatter()** function of matplotlib to plot a scatter plot.


```python
plt.scatter(data['age'], data['income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```

In this example, we are plotting the relationship between the age and income variables in the dataset. We first call the scatter() function and pass the age column as the x-axis values and the income column as the y-axis values. We then label the x and y axes using the xlabel() and ylabel() functions, respectively. Finally, we add a title to the chart using the title() function and display the chart using the show() function.

***Set plot color***

```python
plt.scatter(data['age'], data['income'],color = 'red')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```

***Change edge color of the scatter marker***

```python
plt.scatter(data['age'], data['income'],c = 'red',edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```

***Set size of scatter marker***

```python
plt.scatter(data['age'], data['income'],c = 'red',edgecolor = 'black',s = 100)
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income')
plt.show()
```
